## 1. Basic Python Concepts

```python
# Variables and basic data types
name = "Python Learner"
age = 25
height = 5.8
is_student = True

print(f"Name: {name}")
print(f"Age: {age}")
print(f"Height: {height}")
print(f"Is Student: {is_student}")
```

```python
# Lists and loops
fruits = ["apple", "banana", "orange", "grape", "kiwi"]

print("Fruits list:")
for i, fruit in enumerate(fruits, 1):
    print(f"{i}. {fruit.capitalize()}")

# List comprehension
fruit_lengths = [len(fruit) for fruit in fruits]
print(f"\nFruit lengths: {fruit_lengths}")
```

```python
# Functions
def calculate_bmi(weight, height):
    """
    Calculate BMI (Body Mass Index)
    """
    bmi = weight / (height ** 2)
    return round(bmi, 2)

def bmi_category(bmi):
    """
    Determine BMI category
    """
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal weight"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

# Example usage
weight = 70  # kg
height = 1.75  # meters

bmi = calculate_bmi(weight, height)
category = bmi_category(bmi)

print(f"Weight: {weight} kg")
print(f"Height: {height} m")
print(f"BMI: {bmi}")
print(f"Category: {category}")
```

## 2. Data Analysis with Pandas

```python
import pandas as pd
import numpy as np

# Create sample data
np.random.seed(42)
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Henry'],
    'Age': np.random.randint(20, 60, 8),
    'Salary': np.random.randint(30000, 100000, 8),
    'Department': ['IT', 'HR', 'Finance', 'IT', 'Marketing', 'Finance', 'HR', 'IT'],
    'Experience': np.random.randint(1, 15, 8)
}

df = pd.DataFrame(data)
print("Employee Dataset:")
print(df)
```

```python
# Data exploration
print("Dataset Info:")
print(f"Shape: {df.shape}")
print(f"\nData Types:")
print(df.dtypes)
print(f"\nBasic Statistics:")
print(df.describe())
```

```python
# Data analysis
print("Average salary by department:")
dept_salary = df.groupby('Department')['Salary'].mean().sort_values(ascending=False)
print(dept_salary)

print("\nEmployee count by department:")
dept_count = df['Department'].value_counts()
print(dept_count)

print("\nEmployees with salary > 60000:")
high_earners = df[df['Salary'] > 60000]
print(high_earners[['Name', 'Salary', 'Department']])
```

## 3. Data Visualization with Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
plt.style.use('default')
sns.set_palette("husl")

# Create subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('Employee Data Analysis', fontsize=16, fontweight='bold')

# 1. Salary distribution histogram
axes[0, 0].hist(df['Salary'], bins=6, alpha=0.7, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Salary Distribution')
axes[0, 0].set_xlabel('Salary')
axes[0, 0].set_ylabel('Frequency')

# 2. Department employee count bar chart
dept_count.plot(kind='bar', ax=axes[0, 1], color='lightcoral', alpha=0.8)
axes[0, 1].set_title('Employees by Department')
axes[0, 1].set_xlabel('Department')
axes[0, 1].set_ylabel('Count')
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Age vs Salary scatter plot
scatter = axes[1, 0].scatter(df['Age'], df['Salary'], c=df['Experience'], 
                           cmap='viridis', alpha=0.7, s=100)
axes[1, 0].set_title('Age vs Salary (colored by Experience)')
axes[1, 0].set_xlabel('Age')
axes[1, 0].set_ylabel('Salary')
plt.colorbar(scatter, ax=axes[1, 0], label='Experience (years)')

# 4. Average salary by department
dept_salary.plot(kind='bar', ax=axes[1, 1], color='lightgreen', alpha=0.8)
axes[1, 1].set_title('Average Salary by Department')
axes[1, 1].set_xlabel('Department')
axes[1, 1].set_ylabel('Average Salary')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## 4. Simple Machine Learning Example

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder

# Prepare data for machine learning
# Encode categorical variable
le = LabelEncoder()
df_ml = df.copy()
df_ml['Department_encoded'] = le.fit_transform(df_ml['Department'])

# Features and target
X = df_ml[['Age', 'Experience', 'Department_encoded']]
y = df_ml['Salary']

print("Feature matrix (X):")
print(X.head())
print("\nTarget variable (y):")
print(y.head())
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Model Performance:")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared Score: {r2:.3f}")

# Feature importance (coefficients)
feature_names = ['Age', 'Experience', 'Department']
coefficients = model.coef_

print(f"\nFeature Importance (Coefficients):")
for name, coef in zip(feature_names, coefficients):
    print(f"{name}: {coef:.2f}")
```

```python
# Visualize predictions vs actual values
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.scatter(y_test, y_pred, alpha=0.7, color='blue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Salary')
plt.ylabel('Predicted Salary')
plt.title('Actual vs Predicted Salary')

# Feature importance visualization
plt.subplot(1, 2, 2)
plt.bar(feature_names, coefficients, color=['skyblue', 'lightcoral', 'lightgreen'])
plt.xlabel('Features')
plt.ylabel('Coefficient Value')
plt.title('Feature Importance')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
